**Counting the number of connections (Numberofconnections)**

```python
#counting the number of connections (Numberofconnections) for each unique source IP address (SourceIP) in the UNSWNB15 DataFrame
spark.sql("""
    SELECT
        srcip AS SourceIP,
        COUNT(label) AS Numberofconnections
    FROM
        UNSWNB15
    GROUP BY
        srcip
    ORDER BY
        Numberofconnections DESC
""").show()
```

```python
#Using PySpark to perform a SQL query on our DataFrame named UNSWNB15
df = spark.sql("""
    SELECT srcip AS SourceIP,
           COUNT(label) AS Numberofconnections
    FROM UNSWNB15
    GROUP BY srcip
    ORDER BY Numberofconnections DESC
    LIMIT 10
""").show()
```

```python
#Here, we will generate a Scatter Plot using MatPlotLib to visualize the:
#Relationship between "Source Bytes" (sbytes) & "Destination Bytes" (dbytes) in our PySpark DataFrame UNSWNB15.
# Make sbytes dbytes plot
x = UNSWNB15.select("sbytes").rdd.flatMap(list).collect()
y = UNSWNB15.select("dbytes").rdd.flatMap(list).collect()
# Title, Label axes, set margin, and show plot
plt.title("Sample Visualization")
plt.xlabel('SOURCE Bytes')
plt.ylabel('Destination Bytes')
# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.show()
```

**Hypothesis Testing:**

```python
# Hypothesis Testing: Compare average duration for 'Normal' and 'Attack' connections
from pyspark.sql.functions import avg
avg_duration = UNSWNB15.groupBy('label').agg(avg('dur').alias('avg_duration')).toPandas()

# Visualization: Bar chart for average duration by connection type
avg_duration.plot(kind='bar', x='label', y='avg_duration', legend=False)
plt.title('Average Duration by Connection Type')
plt.xlabel('Connection Type')
plt.ylabel('Average Duration')
plt.show()
```
